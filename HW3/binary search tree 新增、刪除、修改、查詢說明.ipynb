```python
class tree_node(object):
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None
class bst(object):
    def insert(self,root,val):
        if val <= root.val:                  #放入的數值小於或等於原本的值，放左邊
            if root.left:                    #如果左節點存在，與target的比較由root變為root.left(左節點)
                return self.insert(root.left,val)
            else:
                root.left = tree_node(val)   #如果還沒有左節點，val為root的左節點
                return root.left
        else:                                #放入的數值大於原本的值，放右邊
            if root.right:                   #如果右節點存在，與target的比較由root變為root.right(右節點)
                return self.insert(root.right,val)
            else:
                root.right = tree_node(val)  #如果還沒有右節點，val為root的右節點
                return root.right            #回傳value為root的右節點
```

```python
- 新增(放入數值):
1. 首先觀察要放入的值與root的值比較大小
2. 再看原本root的左右邊是否已經有值了

Case1:如果放入的值如果小於或等於原來的root
      放入root的左邊
      如果root的左邊有節點了
      與root左節點比較
      如果沒有其他節點
      放入root的左邊

Case2:放入的值如果大於原來的root
      放入root的右邊
      如果root的右邊有節點了
      與root右節點比較
      如果沒有其他節點
      放入root的右邊
```

```python
class tree_node(object):
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None
class Solution(object):        
    def search(self,root,x):
        if root.val > x:                                       #如果root的值比輸入的值(x)還大
            if (root.left != None) and (root.left.val != x):   #root左節點不為空且不等於x
                return self.search(root.left,x)                #再往左節點遞迴找到與x相同的值
            elif (root.left != None) and (root.left.val == x): #root左節點不為空且等於x
                return root.left                               #回傳位置
            else:                                              #如果沒找到回傳沒有
                return "Fail to find value"
        elif root.val < x:                                     #如果root的值比輸入的值(x)還大
            if (root.right != None) and (root.right.val != x): #root右節點不為空且不等於x
                return self.search(root.right,x)               #再往右節點遞迴找到與x相同的值
            elif (root.right != None and root.right.val == x): #root右節點不為空且等於x
                return root.right                              #回傳位置
            else:                                              #如果沒找到回傳沒有
                return "Fail to find value"
        elif root.val == x:
            return root
```

```python
- 查詢(搜尋數值):
1.判斷值是否有在tree中
2.如果有回傳他對root的位置，沒有則回傳沒找到值

Case 1:如果放入值大於root
       往root的右邊找
       如果root的右邊的值等於放入的值
       回傳root.right(在root的右邊)
       如果右邊的值不等於放入的值
       往右邊的子節點找
       如果右邊沒有值可以找了
       回傳沒找到值
Case 2:如果放入值小於root
       往root的左邊找
       如果root的左邊的值等於放入的值
       回傳root.right(在root的右邊)
       如果左邊的值不等於放入的值
       往左邊的子節點找
       如果左邊沒有值可以找了
       回傳沒找到值  
```

```python
class tree_node(object):
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None
class Solution(object):
    def adjust_tree(self,root):     #調整為符合tree的格式
        current = root
        if current.left != None:    #如果被刪除的節點底下有兩個子節點
            current = current.left  #先往左邊找下一個左節點為新的root
        return current              #回傳現在調整對的root回來
    def preorder(self,root):
        if root:
            print(str(root.val))
        if root.left:
            self.preorder(root.left)
        if root.right:
            self.preorder(root.right)
    def delete(self, root, val):
            #先尋找節點位置
        if val < root.val:
            root.left = self.delete(root.left,val)
        elif val > root.val:
            root.right = self.delete(root.right,val)
        else:
            #要刪除的值底下有一個節點
            if root.left == None:
                new_root = root.right
                root = None
                return new_root
            elif root.right == None:
                new_root = root.left
                root = None
                return new_root
            #要刪除的值底下有兩個節點
            new_root = self.adjust_tree(root.right)
            root.val = new_root.val
            root.right = self.delete(root.right,new_root.val)
        return root
```

```python
- 刪除(移除數值並符合原本構造):
    移除有幾種情況:
    1.刪除值為tree裡沒有，回傳False
    2.刪除值的底下沒有子節點(本身為葉節點)
    3.刪除值底下有一個節點(有左節點或右節點)
    4.刪除值底下有兩個節點
    5.刪除數值後tree也要符合binary search tree的形式
    6.要刪除的值，所有出現在tree中都要刪除
```

```python
class tree_node(object):
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None
class Solution(object):
    def modify(self, root, target, new_val):
        if self.search(root,target) == "Fail to find value":
            return False
        else:
            self.delete(root,target)
            self.insert(root,new_val)
```
